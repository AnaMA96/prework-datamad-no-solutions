```python
from random import choice

gestures = "rock", "paper", "scissors"
print(gestures)

n_rounds = 3

rounds_to_win = (n_rounds/2) + 1

cpu_score = 0
player_score = 0

def returns_random_gestures(no_of_gestures):
    random_gestures = []
    while no_of_gestures > 0:
        random_gestures.append(choice(gestures))
        no_of_gestures -= 1
    return random_gestures
print(returns_random_gestures(3))

def player_gesture():
    print("Please enter your gesture:")
    player_gesture = input()
    while player_gesture not in gestures:
        print ("Please enter one of the following values: 'rock', 'paper' or 'scissors'.")
        player_gesture = input()
    return player_gesture

print("Thank you! You have entered: ", player_gesture())

def the_winner_is(player_gesture, cp_gesture):
    
    if player_gesture == cp_gesture:
        return 0
    if player_gesture == 'rock':
        if cp_gesture == 'scissors':
            return 2
        else:
            return 1
    elif player_gesture == 'paper':
        if cp_gesture == 'rock':
            return 2
        else:
            return 1
    else:
        if cp_gesture == 'paper':
            return 2
        else:
            return 1
        
cpu_score = 0
player_score = 0

def lets_play(player_gesture, cp_gesture):
    global cpu_score
    global player_score
    print ("Computer choice is: ", cp_gesture)
    print ("Player choice is: ", player_gesture)
    winner = the_winner_is(player_gesture, cp_gesture)
    if winner == 1:
        cpu_score += 1
        print ("Computer wins")
    elif winner == 2:
        player_score += 1
        print ("Player wins")

number_rounds = 0
while player_score < rounds_to_win and cpu_score < rounds_to_win and number_rounds < n_rounds:
    player_choice = player_gesture()
    cpu_choice = returns_random_gestures(1)
    lets_play (player_choice, cpu_choice[0])
    number_rounds +=1

if player_score > cpu_score:
    print("Player wins!")
elif player_score < cpu_score:
    print("Cpu wins!")
else:
    print("Tie!")
  
```
